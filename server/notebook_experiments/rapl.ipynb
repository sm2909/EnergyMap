```python
import pytest
import os
import sys
import subprocess
from pyJoules.energy_meter import measure_energy, EnergyContext
from pyJoules.handler.csv_handler import CSVHandler
from tqdm import tqdm

csv_handler = CSVHandler("test.csv")

os.chdir("../black")

class TestCollector:
    def __init__(self):
        self.tests = []

    def pytest_collection_modifyitems(self, session, config, items):
        for item in items:
            self.tests.append(item.nodeid)

def get_tests():
    collector = TestCollector()
    pytest.main(
        ["--collect-only", "-q", "--ignore=tests/test_blackd.py"],
        plugins=[collector],
    )
    return collector.tests

def run_test(test_name, csv_handler):
    with EnergyContext(handler=csv_handler, start_tag=test_name) as ctx:
        subprocess.run(
            [sys.executable, "-m", "pytest", "-q", "--disable-warnings", test_name],
            check=False
        )

def main():
    tests = get_tests()
    # timestamp ; tag(test name) ; duration ; package_energy ; dram_energy
    for test in tqdm(tests[1:20], desc="Running tests"):
        # Pass the handler and test name (as the tag)
        run_test(test, csv_handler)

    csv_handler.save_data()
    print("Data was saved in CSV")

if __name__ == "__main__":
    main()
```
